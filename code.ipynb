```python
import ast
def name(object):
    name=[]
    for value in ast.literal_eval(object):
        name.append(value['name'])
        
    return name    
```

```python
def top_casts(object):
    top_casts=[]
    counter=0
    for value in ast.literal_eval(object):
        if (counter!=3):
            top_casts.append(value['name'])
        else:
            break
        
        counter=counter+1
            
                 
            
    return top_casts        
        
```

```python
data.cast=data.cast.apply(top_casts)
```

```python
def director(object):
    director_list=[]
    for value in ast.literal_eval(object):
        if (value['job']=='Director'):
            director_list.append(value["name"])
            
            
    return director_list        
```

```python
data.crew=data.crew.apply(director)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv=CountVectorizer(max_features=4000,stop_words="english")
```

```python
data.tags=data.tags.apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i))
        
        
    string=" ".join(L)    
    return string
```

```python
data.tags=data.tags.apply(stem)
```

```python
data.head()
```

```python
cv.fit_transform(data.tags).toarray().shape
```

```python
tag_vector=cv.fit_transform(data.tags).toarray()
```

```python
tag_vector
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler(feature_range=(0,1))
numeric_feature_vector=mms.fit_transform(data[['vote_points','revenue','popularity']])
```

```python
numeric_feature_vector
```

```python
final_vector=np.concatenate((tag_vector,numeric_feature_vector),axis=1)
```

```python
final_vector
```

```python
final_vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
similarity=cs(final_vector)
```

```python
def recommend(movie):
    index=data[data.title==movie].index[0]
    movie_list=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(data.iloc[i[0]].title)
```

```python
recommend('Spider-Man 3')
```

```python
recommend('Tangled')
```
